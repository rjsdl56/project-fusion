# 갑자기 뜬금없지만 IoT 데이터를 만들어 이상치를 잡아내보자

```python
import pandas as pd
import random

ax = []
for x in range(101):
    a = random.uniform(-1,1)
    ax.append(a)
    
ay = []
for x in range(101):
    b = random.uniform(-1,1)
    ay.append(b)
    
az = []
for x in range(101):
    c = random.uniform(-1,1)
    az.append(c)
    
print(ax)
print(ay)
print(az)

df = pd.DataFrame([ x for x in zip(ax, ay, az)])
df.columns = ['AX', 'AY', 'AZ']
display(df)
df.to_csv('./data/motion/예상데이터.csv')
```

# 여기서부터는 만들어진 데이터를 가지고 해봅시다.

```python
import pandas as pd
df = pd.read_csv('./data/motion/예상데이터.csv')
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
variance_list = []

for i in range (len(df)):
    try :
        variance = df.iloc[i+1,0] - df.iloc[i,0]
        variance_list.append(abs(variance))
    except :
        variance_list.append(0)
        
df['AX변화량'] = variance_list


variance_list2 = []
for i in range (len(df)):
    try :
        variance2 = df.iloc[i+1,1] - df.iloc[i,1]
        variance_list2.append(abs(variance2))
    except :
        variance_list2.append(0)
        
df['AY변화량'] = variance_list2

display(df)
df.to_csv('./data/motion/예상데이터_변화량.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

sns.set_style('whitegrid')

fig = plt.figure(figsize=(15,10))

sns.boxplot(y='AX변화량', data=df)
plt.show()
```

```python
# 이상치 잡아내기

k = 0
for i in range (len(df)):
    if df.iloc[i,3] > 1.5:
        k += 1
    
print(k)
```

```python
### 군집분석을 해보자 ### 

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import cluster

# # 모형 준비
x = df.iloc[:, :]
x = preprocessing.StandardScaler().fit(x).transform(x)
print(x[:5])

# # 모형 객체 생성
kmeans = cluster.KMeans(init='k-means++', n_clusters=3, n_init=10)

# # 모형 학습
kmeans.fit(x)

# # 예측(군집)
cluster_label = kmeans.labels_
print(cluster_label)

# # 예측결과를 데이터베이스에 입력
df['군집'] = cluster_label
display(df.head(5))

# 그래프로 표현 - 시각화
df.plot(kind='scatter', x='AX변화량', y='AY변화량', c='군집', cmap='Set1', colorbar=False, figsize=(10,5), legend=True)

```
