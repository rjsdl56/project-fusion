```python
import geocoder
from openpyxl import load_workbook
import time
import pandas as pd
```

```python
# 예상데이터 불러오기

data=pd.read_excel("../rider/예상데이터.xlsx")
df=pd.DataFrame(data)
df_address1=df[['출발지']]
df_address2=df[['도착지']]
df['출발지_위도']=''
df['출발지_경도']=''
df['도착지_위도']=''
df['도착지_경도']=''
df['배달거리(km)']=''
#display(df)
```

```python
# 출발지 주소 -> 위도,경도

for i in df_address1.index:

        # geocoder 호출
        g = geocoder.google(df_address1.loc[i], key="AIzaSyDy81EbO46BRSnX1DOgg_F84bhsdbku2z4")

        # 위도, 경도 변환
        geo = g.latlng

        # 변환 확인
        print(geo)

        df['출발지_위도'][i]=geo[0]
        df['출발지_경도'][i]=geo[1]
        
display(df)
```

```python
# 도착지 주소 -> 위도,경도

for i in df_address2.index:

        # geocoder 호출
        g = geocoder.google(df_address2.loc[i], key="AIzaSyDy81EbO46BRSnX1DOgg_F84bhsdbku2z4")

        # 위도, 경도 변환
        geo = g.latlng

        # 변환 확인
        print(geo)

        df['도착지_위도'][i]=geo[0]
        df['도착지_경도'][i]=geo[1]
        
display(df)
```

```python
# 배달거리 측정(km)

from haversine import haversine

for i in df.index :
    
    address1 = (df['출발지_위도'][i],df['출발지_경도'][i]) 
    address2 = (df['도착지_위도'][i],df['도착지_경도'][i])
    distance = haversine(address1, address2, unit = 'km')
    
    df['배달거리(km)'][i] = distance
```
