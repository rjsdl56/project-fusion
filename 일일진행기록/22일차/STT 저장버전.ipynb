# 원본 절대 건들지 말 것!!

```python
# 원본입니다.
# 박성준 버전 저장하기

import io
import os
import pandas as pd

# 제이슨 파일 불러오기? 이건 뭔지 모르겠다 / 인증키 같다
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="./data/test-heracelus-epf9-088e6dc92eba.json"

# 인식할 인텐트 리스트와 라벨링 리스트
intent_list = ['운행시작', '가게전화', '가게도착', '픽업완료', '영수증번호', '소요시간선택', '배달완료']

# 나중에 csv 파일 저장하기 전 시리즈를 저장할 리스트 생성
text = []
intent_index = []
label = []
labeling = -1 # 요건 라벨 인덱스에 들어갈 값 먼저 정해주기

# 인텐트 폴더이름을 for 문으로 돌리자
for intent in intent_list :
    path= "./data/voice/"
    intent_path = intent
    labeling += 1
    print("여기서부터는" + intent + "폴더입니다.")
    
    # 폴더 내에서 파일을 하나씩 돌리자
    for k in range(15): # 각 폴더 중 최대 파일 갯수로 설정
        
        try :
            # 파일명 지정
            file_path = str(k) + ".wav"

            # 구글 클라우드에서 speech 모듈 깔기
            from google.cloud import speech

            # 객체 생성
            client = speech.SpeechClient()

            # 파일 패쓰 지정
            file_name = os.path.join(os.path.dirname(path), intent_path, file_path)

            # 음성 파일을 열어서 텍스트로??
            with io.open(file_name, "rb") as audio_file:
                content = audio_file.read()
                audio = speech.RecognitionAudio(content=content)
            config = speech.RecognitionConfig(
                encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
                ##음성파일 hertz
                sample_rate_hertz=16000, # 나는 m4a를 wav로 변환하니까 16000되서 이렇게 설정
                language_code="ko-KR",
                audio_channel_count=2 # iphone 쓰는 사람은 이렇게 설정해줘야 함
            )

            
            # 텍스트 감지
            response = client.recognize(config=config, audio=audio)

            for result in response.results:
                print(result.alternatives[0].transcript)
                text_value = result.alternatives[0].transcript


#             from hanspell import spell_checker

#             text = result.alternatives[0].transcript

#             sent = text
#             spelled_sent = spell_checker.check(sent)

#             fixed_sent = spelled_sent.checked
            
            # 리스트에 각 저장
            text.append(text_value)
            intent_index.append(intent)
            label.append(labeling)
            
        except :
            pass

# 데이터프레임으로 변환
df = pd.DataFrame([x for x in zip(text, intent_index, label)])
df.columns = ['text', 'intent', 'label']


# 데이터 전처리

# 데이터 확인
display(df)

# 데이터프레임 저장
# df.to_excel('./data/text/박성준녹음.xlsx')
```
